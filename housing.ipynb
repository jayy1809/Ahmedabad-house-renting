MUTUAL INFORMATION

MUTUAL INFORMATION score tells us the relationship between the categorical variables and our target variable , if mi is high means there is high dependency of that categorical variable with target variable

and if its low then there is very less dependency

```python
from sklearn.metrics import mutual_info_score

def calculate_mi(series):
  return mutual_info_score(series, df_train_full.property_price)

df_mi = df_train_full[categorical].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending = False).to_frame(name = 'MI')
df_mi

# hence property_project iand locality are one the most important feature
```

```python
numerical2 =  ['property_sqft', 'num_of_lifts', 'num_of_carparkings', 'num_of_baths']

dict_train = df_train[categorical + numerical2].to_dict(orient = 'records')
```

```python
dict_val = df_val[categorical + numerical2].to_dict(orient = 'records')
```

```python
print(dict_train)
```

```python
from sklearn.feature_extraction import DictVectorizer

dv = DictVectorizer(sparse = False)

X_train = dv.fit_transform(dict_train)
X_val = dv.transform(dict_val)
```

```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
```

```python
xgb_model = XGBRegressor(
    n_estimators=100,  # Number of boosting rounds (you can adjust this)
    learning_rate=0.1,  # Learning rate (you can adjust this)
    max_depth=3,  # Maximum depth of trees (you can adjust this)
    random_state=42  # Random seed for reproducibility
)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
y_pred = xgb_model.predict(X_val)
```

```python
rmse = mean_squared_error(y_val, y_pred, squared=False)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr_model = LinearRegression()

lr_model.fit(X_train, y_train)

y_pred_lr = lr_model.predict(X_val)

rmse_lr = mean_squared_error(y_val, y_pred_lr, squared=False)
print(f"Linear Regression RMSE: {rmse_lr}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

rf_model = RandomForestRegressor(
    n_estimators=100,  # Number of trees (you can adjust this)
    random_state=42    # Random seed for reproducibility
)

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_val)


rmse_rf = mean_squared_error(y_val, y_pred_rf, squared=False)
print(f"RandomForestRegressor RMSE: {rmse_rf}")
```

```python
from sklearn.linear_model import Ridge
```

```python
alpha = 1.0

ridge_model = Ridge(alpha = alpha)

ridge_model.fit(X_train, y_train)
```

```python
y_pred_ridge = ridge_model.predict(X_val)
```

```python
rmse_ridge = mean_squared_error(y_val, y_pred_ridge, squared=False)
print(f"Ridge Regression RMSE: {rmse_ridge}")
```

```python
from sklearn.model_selection import GridSearchCV, cross_val_score
```

```python
rmse_scores = cross_val_score(rf_model, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
rmse_scores = (-rmse_scores) ** 0.5  # Take the square root to get RMSE
mean_rmse = rmse_scores.mean()

print(f"Mean RMSE across {5}-fold cross-validation: {mean_rmse}")
```
